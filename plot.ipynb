```python
import PyPDF2

def pdf_to_txt(pdf_path, txt_path):
    with open(pdf_path, 'rb') as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        num_pages = len(pdf_reader.pages)

        with open(txt_path, 'w') as txt_file:
            for page_num in range(num_pages):
                page = pdf_reader.pages[page_num]
                text = page.extract_text()
                txt_file.write(text)

    print("PDF converted to TXT successfully!")

# Specify the path to your PDF file
pdf_path = '/content/drive/MyDrive/Colab Notebooks/pdf2txt/Eisenhans.pdf'

# Specify the path to save the TXT file
txt_path = '/content/drive/MyDrive/Colab Notebooks/pdf2txt/Eisenhans.txt'

# Call the function to convert PDF to TXT
pdf_to_txt(pdf_path, txt_path)
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Read the text file
file_path = '/content/drive/MyDrive/Colab Notebooks/pdf2txt/Eisenhans.txt'
with open(file_path, 'r', encoding='utf-8') as file:
    text = file.read()

# Tokenize the text into words
tokens = word_tokenize(text)

# Calculate the frequency distribution of words
fdist = FreqDist(tokens)

# Print the most common words and their frequencies
print("Word\t\tFrequency")
print("-----------------------")
for word, frequency in fdist.most_common():
    print(f"{word}\t\t{frequency}")
```

```python
import nltk
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Read the text file
file_path = '/content/drive/MyDrive/Colab Notebooks/pdf2txt/Eisenhans.txt'
with open(file_path, 'r', encoding='utf-8') as file:
    text = file.read()

# Tokenize the text into words
tokens = word_tokenize(text)

# Calculate the frequency distribution of words
fdist = FreqDist(tokens)

# Create a DataFrame from the frequency distribution
data = {'Word': [], 'Frequency': []}
for word, frequency in fdist.most_common():
    data['Word'].append(word)
    data['Frequency'].append(frequency)
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
import nltk
import pandas as pd
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Read the text file
file_path = '/content/drive/MyDrive/Colab Notebooks/pdf2txt/Eisenhans.txt'
with open(file_path, 'r', encoding='utf-8') as file:
    text = file.read()

# Tokenize the text into words
tokens = word_tokenize(text)

# Calculate the frequency distribution of words
fdist = FreqDist(tokens)

# Create a DataFrame from the frequency distribution
data = {'Word': [], 'Frequency': []}
for word, frequency in fdist.most_common():
    data['Word'].append(word)
    data['Frequency'].append(frequency)
df = pd.DataFrame(data)

# Plot the word frequencies
plt.figure(figsize=(20, 6))
plt.plot(df['Word'], df['Frequency'])
plt.xlabel('Word')
plt.ylabel('Frequency')
plt.title('Word Frequency Distribution')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

```

```python
import nltk
import pandas as pd
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

# Read the text file
file_path = '/content/drive/MyDrive/Colab Notebooks/pdf2txt/Eisenhans.txt'
with open(file_path, 'r', encoding='utf-8') as file:
    text = file.read()

# Tokenize the text into words
tokens = word_tokenize(text)

# Calculate the frequency distribution of words
fdist = FreqDist(tokens)

# Get the first 10 most common words and their frequencies
most_common_words = fdist.most_common(10)
words, frequencies = zip(*most_common_words)

# Create a DataFrame from the most common words and frequencies
data = {'Word': words, 'Frequency': frequencies}
df = pd.DataFrame(data)

# Plot the word frequencies
plt.figure(figsize=(10, 6))
plt.plot(df['Word'], df['Frequency'])
plt.xlabel('Word')
plt.ylabel('Frequency')
plt.title('Word Frequency Distribution (Top 10)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
